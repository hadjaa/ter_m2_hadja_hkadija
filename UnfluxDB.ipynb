```python
import csv
import requests
import pandas as pd 
from influxdb_client import InfluxDBClient
from influxdb_client.client.write_api import SYNCHRONOUS
from datetime import datetime

# Create an InfluxDB client instance
influxdb_client = InfluxDBClient(url="http://localhost:8086", token="", org="")

#influxdb_client.switch_database("newdata")
# Open the CSV file
with open("newdata.csv", "r") as file:
    # Create a CSV reader
    reader = pd.read_csv("newdata.csv", sep=",")

    # Initialize a list to store the data
    data = []
    # Set a limit for the number of rows to read from the CSV file
    limit = 1000

    # Keep track of the number of rows processed
    count = 0
    array = []
    time_format = "%H:%M"
    # Loop through each row in the CSV file
    for i,row in reader.iterrows():
        # If the number of rows processed reaches the limit, break the loop
        if count >= limit :
            break
        
        rowsplit = row[0].split(',')
        # Extract the time, pressure, and solar columns from the row
        time = rowsplit[2]
        solar = rowsplit[4]
        presure = rowsplit[5]
        # enlever le '"' du string pour le type time
        time = time.strip('"')

        # Add the data to the list
        data.append({
            "measurement": "douja",
            "tags": {
                "location": "Bresil"
            },
            "time": time,
            "fields": {
                "pressure": float(presure),
                "solar": float(solar)
            }
        })
        
        # Increment the count of rows processed
        count += 1
        
    #influxdb_client.write_api(write_options=SYNCHRONOUS).switch_database("newdata")

    # Write the weather data to the InfluxDB database
    #influxdb_client.write_points(data)
    influxdb_client.write_api(write_options=SYNCHRONOUS).write(bucket="douja", record=data)
    # Print a success message
    print("brezil data stored in InfluxDB successfully!")
```
